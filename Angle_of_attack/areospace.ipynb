# Flight Dynamics and Control ofAerospace Vehicles
'''Flight dynamics deals principally with the response of aerospace vehicles to perturbations
in their flight environments and to control inputs. In order to understand this response,
it is necessary to characterize the aerodynamic and propulsive forces and moments acting
on the vehicle, and the dependence of these forces and moments on the flight variables,
including airspeed and vehicle orientation. These notes provide an introduction to the
engineering science of flight dynamics, focusing primarily of aspects of stability and
control. The notes contain a simplified summary of important results from aerodynamics
that can be used to characterize the forcing functions, a description of static stability
for the longitudinal problem, and an introduction to the dynamics and control of both,
longitudinal and lateral/directional problems, including some aspects of feedback control.'''

```python
#import libararies that we use 
import math 
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline 
```

```python
#Allow the user to input value of veocity 
u= float(input('please enter your velocity in roll axis = '))
v= float(input('please enter your velocity in pitch axis = '))
w= float(input('please enter your velocity in yaw axis = '))
```

```python
#velocity vector 
V = np.sqrt(u**2 + v**2 + w**2)
```

```python
#calculate angles With knowledge u,v,w
angle_of_attack = math.tanh(w/u)
sideslip_angle = math.sinh(v/V)
```

```python
#transfer angle from radian to degree
angle_of_attack = math.degrees(angle_of_attack)
sideslip_angle = math.degrees(sideslip_angle)
```

```python
#check the output
print(angle_of_attack,'km/hr','\n' ,sideslip_angle,'km/hr')
```

# plot angle_of_attack and sideslip_angle with With knowledge u,v,w 
## u is varing from 500 to 600 km/hr
## v is varing from 0 to 50 km/hr
## w is varing from 0 to 100 km/hr

```python
u = np.arange(500,600)
v = np.arange(0,50,0.5)
w = np.arange(0,100)


V = np.sqrt(u**2 + v**2 + w**2)


angle_of_attack = np.tanh(w/u)
sideslip_angle = np.sinh(v/V)

angle_of_attack = np.degrees(angle_of_attack)
sideslip_angle = np.degrees(sideslip_angle)
plt.figure(figsize=(10,10))
plt.title('angle attack & sideslip')
plt.xlabel('the velocity in roll axis')

plt.plot(V,angle_of_attack)
plt.plot(V,sideslip_angle)
plt.legend(('angle_of_attack', ' sideslip_angle'))

```
